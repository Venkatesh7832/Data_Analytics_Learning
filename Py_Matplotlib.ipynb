```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use('dark_background')
```

```python
#A library for data visualisation
roll = [1,2,3,4,5,6,7,8,9]
marks = [10,20,30,40,50,60,70,80,90]

plt.figure(figsize=(8,8))
plt.scatter(roll,marks,label='Roll No Vs Marks',marker='*',s=25)
plt.title('Roll No Vs Marks')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.legend()
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
print(end="<------------------------------------------------------------------------------------------------------------------>")
plt.figure(figsize=(8,8))
plt.plot(roll,marks,'ro',markersize=15)
plt.title('Roll No Vs Marks')
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.show()
```

```python
# Two graphs in one-plot:
temperature = np.array([28,26,22,27,35,37,30,31,25,38,24,34])
humidity = np.array([60, 62, 61, 59, 60, 58, 55, 63, 65, 60, 59, 57])

plt.figure(figsize=(10,6))
plt.xticks(np.arange(20, 40, 2))
plt.yticks(np.arange(50, 70, 2))
print('______________________________________>')
plt.scatter(temperature, humidity, marker='o', s=100, label='Temperature vs Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Humidity (%)')
plt.title('Temperature vs Humidity Scatter Plot')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

''' # Assume our data is in a CSV file named 'weather_data.csv'
    # with columns like 'Date', 'Temperature', 'Humidity'
    weather_data = pd.read_csv('weather_data.csv')
    temperatures = weather_data['Temperature']
    humidities = weather_data['Humidity']
    # Now we plot the graphs
'''
```

```python
# Line Plot :
roll_no = [1, 2, 3, 4, 5, 6, 7]
marks = [32, 26, 43, 41, 16, 37, 29]

plt.plot(roll_no,marks,'r-')  # linestyles = '-','--','-.',':' ..etc
plt.xlabel('Roll No')
plt.ylabel('Marks')
plt.title('Roll No Vs Marks')
plt.show()
print("Multple plots in a graph")
plt.figure(figsize=(8,8))
plt.xticks(np.arange(0,8,1))
plt.yticks(np.arange(12,48,4))
plt.plot(roll_no,marks,'r-')
plt.plot(roll_no,marks,'b.')
plt.xlabel('Roll Numbers')
plt.ylabel('Marks')
plt.title('Roll No Vs Marks')
#plt.legend()
plt.grid(True,linestyle='--',alpha=0.8)
plt.show()
```

```python
# Bar Graph
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]

plt.bar(fruits, sales)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()

print("<------------------------Graph with different colours-------------------------->")
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
sales = [400, 350, 300, 450]
colour = ['blue','green','red','white']

plt.bar(fruits, sales,color=colour)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Sales')
plt.show()
print("<------------------------Horizontal Bar Graph-------------------------->")
numbers = [10,20,30,40,50,60,70]
five_mul = [5,10,15,20,25,30,35]

plt.barh(numbers, five_mul)
plt.title('Number Vs Five Multiples')
plt.xlabel('Numbers')
plt.ylabel('Five Multiples')
plt.show()
print("<------------------------Two plots in a Graph-------------------------->")
subjects = ['English','Maths','Physics','Chemistry']
marks1 = [91,99,83,92]
marks2 = [85,97,86,98]

plt.figure(figsize=(8,8))
bar_width = 0.35
r1 = np.arange(len(subjects))
r2 = [x + bar_width for x in r1]

plt.bar(r1, marks1, width=bar_width, label='Marks 1',color='White')
plt.bar(r2, marks2, width=bar_width, label='Marks 2',color='Green')

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Subjects Vs Marks')
plt.xticks([r + bar_width/2 for r in range(len(subjects))], subjects) # Set the x-tick labels
plt.legend()
plt.show()
```
