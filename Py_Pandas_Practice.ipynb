```python
# LOC & iLOC
print("loc:")
lst = [1,2,3,4,5,6,7,8,9,0,[9,3,4,5]]
df = pd.DataFrame(lst)
print(df)
print(df.loc[1])
print(df.loc[3:5])
print("Iloc:")
print(df.iloc[5])
print(df.iloc[3:7])
print(df.iloc[1:4,5:7])
```

```python
# Groupby, Merge and Append Functions
#Group-by
data = {'Region': ['North', 'South', 'North', 'East', 'South', 'West', 'North', 'East'],'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
           'Sales': [100, 150, 120, 80, 200, 90, 110, 130],'Quantity': [10, 15, 12, 8, 20, 9, 11, 13]}
df = pd.DataFrame(data)
print(df)
print(df.groupby('Region').sum())
print(df.groupby(['Region','Product']).sum()) # Fixed: Used parentheses instead of square brackets

#Merge->
data1 = {'employee_id': [1, 2, 3, 4, 5],'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],'department': ['HR', 'IT', 'Finance', 'HR', 'IT']}

data2 = {'employee_id': [2, 3, 5, 6, 7],'salary': [60000, 75000, 80000, 50000, 65000],
          'start_date': ['2020-01-15', '2019-03-01', '2021-07-20', '2022-05-10', '2023-09-01']}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
merged_df = pd.merge(df1, df2, on='employee_id', how='inner') # Performed merge and assigned to merged_df
print("\nMerged DataFrame:")
print(merged_df)


#Append->
d1 = {'Name': ['Alice', 'Bob', 'Charlie'],'Age': [25, 30, 35],'City': ['New York', 'Los Angeles', 'Chicago']}

d2 = {'Name': ['David', 'Eve'],'Age': [40, 28],'City': ['Houston', 'Miami']}
df3 = pd.DataFrame(d1)
df4 = pd.DataFrame(d2)
print(df3)
print(df4)
appended_df = pd.concat([df3, df4], ignore_index=True)
print(appended_df)
```

```python
# Pivot Table
data = {'Region': ['North', 'South', 'North', 'East', 'South', 'West', 'North', 'East'],'Product': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'B'],
           'Sales': [100, 150, 120, 80, 200, 90, 110, 130],'Quantity': [10, 15, 12, 8, 20, 9, 11, 13]}
df = pd.DataFrame(data)
print(df)
print("Pivot Table:")
pivot_table = df.pivot_table(values='Sales',index='Region',columns='Product',aggfunc='sum')
print(pivot_table)
pivot_table = df.pivot_table(values='Sales',index='Region',columns='Product',aggfunc='mean')
print(pivot_table)
pivot_table = df.pivot_table(values='Sales',index='Region',columns='Product',aggfunc='count')
print(pivot_table)
pivot_table = df.pivot_table(values='Sales',index='Region',columns='Product',aggfunc='max')
print(pivot_table)
```
